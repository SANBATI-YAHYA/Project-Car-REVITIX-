```python
import joblib
from sklearn.ensemble import IsolationForest

def detect_outlier_isolation_forest(marque_input, poids_input, df, contamination=0.07):
    # ŸÜŸÅŸäÿ™ÿ±ŸäŸà ÿ∫Ÿäÿ± ÿßŸÑÿ®ŸäÿßŸÜÿßÿ™ ÿØŸäÿßŸÑ ÿ∞ŸäŸÉ marque
    subset = df[df['Manufacturer'] == marque_input]

    # ŸÜÿ™ÿ£ŸÉÿØŸà ÿ®ÿßŸÑŸÑŸä ÿπŸÜÿØŸÜÿß ÿπÿØÿØ ŸÉÿßŸÅŸä ŸÖŸÜ les poids
    if len(subset) < 5:
        return f"üö´ Not enough data for marque '{marque_input}' ({len(subset)} rows)"

    # ŸÜŸÖŸàÿ∞ÿ¨ Isolation Forest
    model = IsolationForest(contamination=contamination, random_state=42)
    model.fit(subset[['Weight_Avg']])

    # ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ÿ®ÿπÿØ ÿßŸÑÿ™ÿØÿ±Ÿäÿ® ŸÖÿ®ÿßÿ¥ÿ±ÿ©
    joblib.dump(model, f'isolation_forest_{marque_input}_model.pkl')  # ÿ≠ŸÅÿ∏ ÿßŸÑŸÜŸÖŸàÿ∞ÿ¨ ŸÅŸä ŸÖŸÑŸÅ

    # Prediction
    prediction = model.predict([[poids_input]])

    if prediction[0] == -1:
        return f"‚ùå {poids_input} KG is an **outlier** for '{marque_input}'"
    else:
        return f"‚úÖ {poids_input} KG is **normal** for '{marque_input}'"
```

```python
result = detect_outlier_isolation_forest("BMW", 2100, df)
print(result)
```

```python
detect_outlier_isolation_forest("NISSAN", 2300, df)
```

```python
import joblib
from sklearn.ensemble import IsolationForest

def train_and_save_models(df, contamination=0.07):
    marques = df['Manufacturer'].unique()
    for marque in marques:
        subset = df[df['Manufacturer'] == marque]
        if len(subset) >= 5:
            model = IsolationForest(contamination=contamination, random_state=42)
            model.fit(subset[['Weight_Avg']])
            joblib.dump(model, f'isolation_forest_{marque}.pkl')
            print(f"‚úÖ Model saved for {marque}")
        else:
            print(f"‚õî Not enough data for {marque}")
```

```python
train_and_save_models(df)
```
